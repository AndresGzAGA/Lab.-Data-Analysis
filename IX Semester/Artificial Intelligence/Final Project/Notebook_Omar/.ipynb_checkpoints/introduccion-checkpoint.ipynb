### Instalación de paquetes

   <font size=4 color='blue'>
    
   Tensorflow es compatible con los siguientes sistemas de 64 bits:

    
        Python 3.5 a 3.8
        Ubuntu 16.04 o versiones posteriores
        Windows 7 o versiones posteriores
        macOS 10.12.6 o versiones posteriores
        Raspian 9.0 o versiones posteriores
    
Para instalar Tensorflow es posible hacerlo mediante el administrador de paquetes *pip* de Python, ahora se procede a explicar la manera de instalarlo en la distribucion Linux.
    
Abrimos una terminal y verfificamos la version de Python existente con
    
    $ python3 --version
    
En caso de no contar con alguna version compatible con tensorflow realizamos lo siguiente. Primeramente realizamos una actualizacion de software con el comando
    
    $ sudo apt-get upgrade
    
Posteriormente instalamos Python con el comando 
    
    $ sudo apt-get install Python 3.x.x (sustituyendo la version deseada)
    
Con esto ya contaremos con la herramienta *pip*, procedemos a instalar Tensorflow siguiendo los siguientes comandos
    
    $ pip install --upgrade pip
    
    $ pip install tensorflow

### Ventajas y desventajas de este framework

<font size=4 color='blue'>
    
Tensorflow es un framework que funciona por medio del flujo de tensores, los cuales son arreglos n-dimensionales, estos son operados mediante graficos computacionales, en este paradigma existe una red de nodos donde cada nodo realiza una operacion matematica, de esta manera se pueden manejar datos de multiples tipos.
En Tensorflow se pueden construir distintos algoritmos de deep learning por ejemplo CNN, RNN, DBN, NLP, etc. y existe una afinidad especial para el trabajo de machine learning.
    
Tensorflow utiliza Python para su API front-end siendo conveniente para el desarrollo de aplicaciones con el framework mientras ejecuta las aplicaciones en C++ obteniendo asi un mayor rendimiento. Es por todo esto que existen ventajas y algunas deventajas en el uso de este framework enunciando algunas a continuacion.

| Ventajas | Desventajas |
| --- | --- |
|Cuenta con API's de alto y bajo nivel proporcionando asi un manejo mas sencillo y un buen desempeño | No es tan sencillo de utilizar a pesar de proporcionar API's de alto nivel
|Su biblioteca es compatible con varios lenguajes de programacion como Python, C++, java. | Su para proyectos cortos o con poco tiempo de preparacion resulta mas simple con otros frameworks |
|Es portable y se puede utilizar en todo tipo de dispositivos, ademas de contar con el respaldo de google por lo cual cuenta con un desarrollo rapido y mejoras para una implementacion y uso mas sencillos.|  |
|Cuenta con un rendimiento acelerado en la nube de google asi como un hub online para compartir proyectos|  |

### Particularidades de este framework

<font size=4 color='blue'>
    
Como se mencionó antes, Tensorflow es un framework con un alto rendimiento debido a su ejecucion en c++ de alto rendimiento y solo controlando el trafico etre piezas por medio de python para proporcionar abstracciones de alto nivel para unirlas, por lo cual resulta conveniente en proyectos con grandes conjuntos de datos que necesitan una ejecucion rapida, por otro lado existen frameworks mas amigables para el desarrollo de proyectos cortos o con un menor tiempo de preparación.
    
Ya que es compatible con distintos lenguajes de programación y sistemas operativos, su uso es posible ya sea desde utilizarlo desde el navegador web hasta en ordenadores como el raspberry pi con enfoques tanto en mejor rendimiento como en una implementacion mas sencilla.
    
La flexibilidad del framework permite a desarrolladores y cientificos porbar nuevas ideas en un tiempo corto además miles de usuarios tienen la posibilidad de utilizar proyectos desarrollados en Tensorflow.
