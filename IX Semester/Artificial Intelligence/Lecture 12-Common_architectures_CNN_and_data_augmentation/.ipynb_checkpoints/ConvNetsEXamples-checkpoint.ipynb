 # ConvNets

<font size=4>

Una arquitectura de red neuronal que funciona bien en una tarea de visión por computadora a menudo funciona bien en otras tareas.

A continuación se revisara la arquitectura de tres CNN muy útiles. 

<font size=4 color='steelblue'>
    
> # 1. LeNet

<font size=4>
    
 Toma su nombre del principal autor del artículo en el cual se publicó: LeCun. 

El objetivo de LeNet-5 era reconocer los dígitos escritos a mano. 

LeNet-5 fue entrenado en imágenes en escala de grises, por lo que la entrada tiene dimensiones 32 por 32 por 1.

Esta red neuronal es pequeña para los estándares modernos.


A medida que profundizas en una red, la altura y la anchura tienden a disminuir.


El patron general a seguir es colocar una capa convolucional seguida de una capa de agrupamiento, repetir lo anterior y después conectar con capas totalmente conectadas hasta llegar a las salidas.

En la siguiente imagen se muestra la arquitectura de la red LeNet-5:


<img src='https://drive.google.com/uc?id=1hDOR1OoAeRKs8QMPA223B8Fv2WlYf7_i' style="width: 900px;"/>


En particular, la arquitectura de la imagen anterior se denomina LeNet-5 porque tiene 5 capas.

> # 3. VGG

<font size=4>

En lugar de tener tantos hiperparámetros, utiliza una red mucho más simple donde se centra en tener capas convolucionales con filtros de tres por tres, con un paso de uno y siempre utilizan padding de tipo *same*.

Todas sus capas de agrupación máxima son de tamaño dos por dos con un paso de dos. 


 <img src='https://drive.google.com/uc?id=10t7o6Z0pXgDE2qBAFXmuW273GfuCGxeN' style="width: 900px;"/>


El número de parámetros entrenables para esta arquitectura es de 138 millones.

Se denomina VGG-16 porque tiene 16 capas.
