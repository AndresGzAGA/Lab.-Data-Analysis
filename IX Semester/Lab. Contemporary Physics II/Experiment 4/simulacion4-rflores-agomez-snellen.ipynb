```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#declaracion de parametros del experimento

ni = 800 #numero de incidencias
ne = 100 #numero de experimentos
phi0 = np.pi/4 #fase en grados
theta0 = np.pi/4 #polarizacion inicial
sang = np.pi/120 #error en los angulos de los polarizadores
```

```python
def bell_m(ni, phi0, theta0, pang, sn = 20, stheta = np.pi/72, sphi = np.radians(2.5)):
    a = pang[0]
    b = pang[1]
    #nf = np.int(nf)
    #se introducen incertidumbres experimentals
    theta = np.pi/4 + rand.normal(0, stheta) #desfase de las polarizaciones
    n = np.int(ni + rand.normal(0, sn)) #introduce incertidumbre en el numero de incidencias
    phi = rand.normal(phi0, sphi) #genera la fase del experimento
    
    #se realiza el conteo
    amp = np.sin(a)*np.sin(b)*np.cos(theta) + np.exp(1j*phi)*np.cos(a)*np.cos(b)*np.sin(theta)
    N = n*amp*np.conj(amp) + rand.normal(0.03*n, sn)
    return np.int(np.real(N))
```

```python
def bell_0(angarray, ni, phi0, theta0, sn = 20, stheta = np.pi/72, sphi = np.radians(2.5)):
    #angarray debe ser el array de 2x16 angulos
    med = np.array([bell_m(ni, phi0, theta0, aaa, sn, stheta, sphi) for aaa in angarray])
    sums = [np.sum(med[4*j:4*j+4]) for j in range(4)]
    nums = [np.sum(med[4*j:4*j+2]) - np.sum(med[4*j+2:4*j+4]) for j in range(4)]
    E = [nums[j]/sums[j] for j in range(4)]
    S = E[0] - E[1] + E[2] + E[3]
    return S
```

```python
def bell(ne, ni, phi0, theta0, sn = 20, stheta = np.pi/72, sphi = np.pi/120, sang = np.pi/120):
    ne = np.int(ne)
    eses = np.empty(ne)
    for i in range(ne):
        ang = pd.read_csv('angbell.dat', delim_whitespace=True, header = None) #lee y convierte angbell en un array
        ang = ang.to_numpy()
        ang = np.radians(ang)
        ang = np.array([j + rand.normal(0, sang) for j in ang]) #introduce error en los angulos 
        
        eses[i] = bell_0(ang[0:16], ni, phi0, theta0, sn, stheta, sphi)
    eses = pd.DataFrame(eses)
    return eses.describe()
```

```python
bell(100, ni, phi0, np.pi/4 )
```
