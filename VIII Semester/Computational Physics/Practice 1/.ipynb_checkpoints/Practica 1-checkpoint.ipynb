### 1. Conversión entre numeros decimales y números binarios

```python
def binaria(N):
    B = []
    while True:
        B.append(N%2)
        N = int(N/2)
        if N == 0:
            break
    B.reverse()
    return B
```

```python
binaria(29)
```

```python
def decimal(L):
    L.reverse()
    N = 0
    for i in range(len(L)):
        N += L[i]*2**(i)
    return N
```

```python
decimal([1,1,1,0,1])
```

### 4. Guerra de métodos

```python
def Incremental(f,a,b,dx=1e-5):
    i = 0
    for x in np.arange(a,b,dx):
        if f(x)*f(x+dx) <= 0:
            return [x,i]
        i+=1
    print('Raiz no encontrada')
    return [a,i]

def Biseccion(f,a,b,E=1e-6):
    i = 0
    if f(a)*f(b) >= 0:
        print('Error, tienes que elegir un intervalo con cambio de signo')
        return [np.nan,i]
    x1 = a
    x2 = b
    while abs(x2-x1) > E:
        x3 = 0.5*(x2+x1)
        if f(x1)*f(x3) < 0:
            x2 = x3
        elif f(x2)*f(x3) < 0:
            x1 = x3
        elif f(x3) == 0:
            return [x3,i]
        i+=1
    return [x3, i]

def Newton(f,df,x,E=1e-6):
    i=0
    x0 = x
    while abs(f(x0)) > E:
        x1 = x0 -f(x0)/df(x0)
        x0 = x1
        i += 1
    return [x0,i]

def Secante(f, x0, x1, E=1e-6):
    i=0
    while abs(f(x0)) > E:
        x2 = x1 - f(x1)*((f(x1)-f(x0))/(x1-x0))**(-1)
        x0 = x1
        x1 = x2
        i+=1
    return [x1,i]

def Ridders(f, x0, x2, E  = 1e-6):
    i=0
    if f(x0)*f(x2) >= 0:
        print('Intervalo inválido, elige uno en donde haya un cambio de signo')
        return [np.nan,0]
    while True:
        x1 = 0.5*(x0+x2)
        x3 = x1 + np.sign(f(x0))*(x2-x1)*f(x1)/np.sqrt(f(x1)**2-f(x0)*f(x2))
        if abs(f(x3)) < E:
            break
        elif f(x3)*f(x0) < 0:
            x2 = x3
        elif f(x3)*f(x1) < 0:
            x0 = min(x3,x1)
            x2 = max(x3,x1)
        elif f(x3)*f(x2) < 0:
            x0 = min(x3,x2)
            x2 = max(x3,x2)
        else:
            print('Algo salió mal')
        i+=1
    return [x3,i]
```

```python
f1 = lambda x: 2*x**2+5*x-3 #Dos raíces
df1 = lambda x: 4*x+5

x = np.arange(-7,7,0.01)
plt.plot(x,f1(x))
plt.grid()
plt.show()
```

2 intervalos necesarios para cada raíz, ambos en los cuales debe haber cambio de signo

```python
#Primer intervalo para encontrar el cero cercano a -3
a = -3.5
b = -2
print(Incremental(f1, a, b, dx = 1e-5))
print(Biseccion(f1, a, b, E=1e-6))
print(Newton(f1, df1, 0.5*(a+b), E=1e-6))
print(Secante(f1, a, b, E=1e-6))
print(Ridders(f1, a, b, E=1e-6))
```

```python
#Primer intervalo para encontrar el cero cercano a 0
a = -0.5
b = 2
print(Incremental(f1, a, b, dx = 1e-5))
print(Biseccion(f1, a, b, E=1e-6))
print(Newton(f1, df1, 0.5*(a+b), E=1e-6))
print(Secante(f1, a, b, E=1e-6))
print(Ridders(f1, a, b, E=1e-6))
```

En ambos casos Ridders lo encuentra en menos pasos

### 5. Movimiento de proyectil con fricción

```python
Y = lambda t: 4605*(1-np.exp(-t/15))-147*t
X = lambda t: 2400*(1-np.exp(-t/15))

t = np.arange(0,30)
plt.plot(X(t),Y(t))
plt.grid()
plt.show()
```

Podemos encontrar el tiempo en el que Y(t) se hace cero con Ridders, solo hay que tantear el intervalo del tiempo de impacto

```python
print(Ridders(Y, 20, 30, E=1e-10))
```
