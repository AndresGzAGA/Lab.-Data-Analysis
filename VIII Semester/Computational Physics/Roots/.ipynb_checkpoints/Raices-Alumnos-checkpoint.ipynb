# Raíces de funciones

Buscamos los valores para $x$ donde $f(x) = 0$ para una función $f$ conocida. Estos valores se llaman *ceros* o *raíces* de $f$ y se denotan $\alpha$.

```python
import numpy as np
import pandas as pd
```

# Método de Ridders

```python
#f = lambda x: -24+26*x-9*x**2 + x**3  #Es (x-2)*(x-3)*(x-4)
f = lambda x: x**3-x-5

def Ridders(f, x0, x2, E  = 10**(-6)):
    if f(x0)*f(x2) >= 0:
        print('Intervalo inválido, elige uno en donde haya un cambio de signo')
        return [np.nan,0]
    i = 0
    while True:
        i += 1
        x1 = 0.5*(x0+x2)
        x3 = x1 + np.sign(f(x0))*(x2-x1)*f(x1)/np.sqrt(f(x1)**2-f(x0)*f(x2))
        if abs(f(x3)) < E:
            break
        elif f(x3)*f(x0) < 0:
            x2 = x3
        elif f(x3)*f(x1) < 0:
            x0 = min(x3,x1)
            x2 = max(x3,x1)
        elif f(x3)*f(x2) < 0:
            x0 = min(x3,x2)
            x2 = max(x3,x2)
        else:
            print('Algo salió mal')
    return [x3,i]
```

```python
raiz, iteraciones = Ridders(f,1.2,2.9)
print(raiz)
print(f(raiz))
print(iteraciones)
```

```python
lista_incremental = []
idx = []
valor_real=1.904160859
for i in range(1,6):
    tol = 10**(-i)
    idx.append(tol)
    raiz=Ridders(f,-1,3,tol) #cada renglon de la tabla
    error=(abs(valor_real-raiz[0])/valor_real)*100
    lista_incremental.append([raiz[0],raiz[1],error])
cols = ["Raiz","Pasos","error %"]
pd.DataFrame(lista_incremental, index=idx, columns=cols)
```
