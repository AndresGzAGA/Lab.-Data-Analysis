4. Sucesión de números aleatorios en el intervalo [0.01, 0.20] cuya suma sea menor a 1.50

```python
S = []
x = np.random.random(1)[0]/5.0 #Numero aleatorio en el intervalo [0.0,1.0] dividido entre 5 estará [0.0,0.20]
while sum(S) + x < 1.50:
    if x >= 0.01:
        S.append(x)
    x = np.random.random(1)[0]/5.0
print(S)
print(sum(S))
```

6. Potencial escalón Cuántico

```python
#Energía en eV
E = 10.0
V = 9.0
#Masa en kg x 10^{-31}, los factores no importan en las fórmulas de T y R, con tal de que seamos consistentes
m = 9.11
#De igual manera, h barra es un factor que no cambia las fórmulas para T y R, podemos omitirlo
k1 = np.sqrt(2*m*E)
k2 = np.sqrt(2*m*(E-V))

T = 4*k1*k2/(k1+k2)**2
R = ((k1-k2)/(k1+k2))**2

print('T = '+str(T))
print('R = '+str(R))
```

8. La fórmula semi-empírica de la masa

(i) Pedir los valores de A y Z al usuario e imprimir B

```python
a1 = 15.8
a2 = 18.3
a3 = 0.714
a4 = 23.2

while True:
    try:
        A = float(input("A = "))
        break
    except ValueError:
        print("Ese no es un número. Intenta de nuevo")
        
while True:
    try:
        Z = float(input("Z = "))
        break
    except ValueError:
        print("Ese no es un número. Intenta de nuevo")
        
if A%2 != 0: #A no es divisible por 2: Impar
    a5 = 0.0
elif Z%2 == 0: #Aquí A ya es par, solo debemos checar si Z lo es
    a5 = 12.0
else: #Aquí entonces ambos son impares
    a5 = -12.0

B = a1*A-a2*A**(2/3)-a3*Z**2/(A**(1/3))-a4*(A-2*Z)**2/A+a5/A**(1/2)
print ('B = '+str(B))
```

9. Constante de Madelung

```python
L = 200
# V = e*M/(4*pi*Eo*a), entonces solo nos interesa calcular (4*pi*Eo*a)*V/e = +- 1/(i**2+j**2+k**2)^(1/2)

def V_en(i,j,k,n):
    'Te da la suma correspondiente dado el número de veces que ya tomas en cuenta ese sector'
    if (i+j+k) % 2 == 0: #Carga positiva
        return n/(i**2+j**2+k**2)**(1/2)
    else: #Carga negativa
        return -n/(i**2+j**2+k**2)**(1/2)

M = 0
#Nos tomamos 1/2 del espacio
for i in range(1, L+1):
    #Tomamos 1/2 de cada plano
    for j in range(1, L+1):
        #Tomamos 1/2 de cada recta (8 pedazos en total)
        for k in range(1, L+1):
            M += V_en(i,j,k,8) #8 veces porque tomamos en cuenta i y -i, j y -j, k y -k

#Completamos con los planos coordenados x-y, y-z, x-z que habíamos ignorado, pues no se toman 2 veces          
for j in range(1, L+1): #Nos tomamos 1/2 del plano
    for k in range(1, L+1):
        M += V_en(0,j,k,12) #3planos*4 veces porque tomamos en cuenta j y -j, k y -k

#Completamos con los ejes x,y,z que habíamos ignorado
for k in range(1, L+1):
    M += V_en(0,0,k,6) #3planos*2 veces porque tomamos en cuenta k y -k

#El origen no se toma en cuenta

print('M = '+str(M))

#Se tarda un poco para calcular con L=200, pero no tanto como  cuando uno corre todos los índices en todo el espacio
```

10. Tiro parabólico

```python
print('Introduce la velocidad inicial en m/s')
while True:
    try:
        v = float(input("v = "))
        break
    except ValueError:
        print("Ese no es un número. Intenta de nuevo")

def y(t):
    return v*t-9.81*t**2/2

H = 0
t = 0
while y(t) >= 0:
    t += 0.1
    if y(t) > H:
        H = y(t)
print('La altura máxima es: '+str(H))
```

12. Matriz aleatoria simétrica

```python
n=5
A = np.zeros((n,n))

for i in range(0,n):
    for j in range(i,n):
        x = np.random.random(1)[0]/5.0 #Numero aleatorio en [0,2.0]
        A[i][j] = x
        A[j][i] = x #Matriz simétrica
print(A)
```

13. Matriz de Vandermonde

```python
print('Introduce el tamaño del vector')

while True:
    try:
        n = int(input("n = "))
        break
    except ValueError:
        print("Ese no es un número. Intenta de nuevo")

print('Introduce las '+str(n)+' componentes del vector')
V = np.zeros(n)
for i in range(0,n):
    while True:
        try:
            V[i] = float(input('V'+str(i)+" = "))
            break
        except ValueError:
            print("Ese no es un número. Intenta de nuevo")

A = np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        A[j][i] = V[j]**i
print(A)
```

14. Matriz tridiagonal

```python
print('Introduce el tamaño del vector')

while True:
    try:
        n = int(input("n = "))
        break
    except ValueError:
        print("Ese no es un número. Intenta de nuevo")

print('Introduce las '+str(n)+' componentes del vector')
V = np.zeros(n)
for i in range(0,n):
    while True:
        try:
            V[i] = float(input('V'+str(i)+" = "))
            break
        except ValueError:
            print("Ese no es un número. Intenta de nuevo")

A = np.zeros((n,n))
A[0][0] = V[0] #Component de la esquina superior izquierda, que no tomaremos en cuenta en el for
for i in range(1,n):
    A[i][i] = V[i]
    A[i-1][i] = V[i]**2
    A[i][i-1] = V[i]**3
print(A)
```
