```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#import seaborn as sns
from scipy.optimize import curve_fit
from pandas.tools.plotting import table
from mpl_toolkits.mplot3d import Axes3D


%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 9] # bigger figures
plt.rcParams['font.size'] = 20
pd.set_option('display.max_rows',100)
```

```python
# Separtors: whitspace, not just tab
# Don't use first column as index
# more info: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_table.html
dat = pd.read_table(open('ln.txt'), delim_whitespace=True, index_col=False)
```

```python
dat
```

```python
dat.d = np.abs(dat.d1 + dat.d2)/2
dat.si = dat.sisod - dat.so - dat.d
dat.ssd = np.abs(dat.d1 -dat.d2)/2
dat.ssi = np.sqrt(dat.ssd**2 + 2*0.1**2)
```

```python
dat.describe()
```

```python
def f(so, si):
    return si*so/(si+so)
```

```python
def SI(SO, F):
    return SO*F/(SO-F)
SI(14,7)
```

```python
fit_res = curve_fit(SI, dat.si, dat.so, sigma = dat.ssi, p0 = 7)
print(fit_res)
fg = fit_res[0] 
sg=(np.diag(fit_res[1]))
print(fg)
```
