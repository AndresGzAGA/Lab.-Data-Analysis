```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inicialización de variables. Las variables se deben definir con un valor inicial predeterminado
w = tf.Variable(tf.random.normal((3, 2)), name='w') #Matriz con shape (3,2) con dist normal alrededor del 0
b = tf.Variable(tf.zeros(2, dtype=tf.float32), name='b') #Vector de ceros de longitud 2
x = [[1., 2., 3.]] #Lista usual de python
```

```python
print("w = "+str(w.numpy())) #w es el objeto, y w.numpy() es el valor del objeto, representado en un arreglo de numpy
print("b = "+str(b.numpy()))
print("x = "+str(x))
```

```python
#Ejemplo de operación multi-lineal
y = x @ w + b
loss = tf.reduce_mean(y**2) #Promedio de los cuadrados (función de costo)

print("y = "+str(y.numpy()))
print("loss = "+str(loss.numpy()))
```

## Automatic differentiation

```python
#Diferenciación de y con respecto a las variables w y b

#Tape registra las operaciones que se realizan entre las variables, análogo a definir una función
with tf.GradientTape(persistent=True) as tape:
  y = x @ w + b
  loss = tf.reduce_mean(y**2)
    
#Con esto entonces se puede hacer la diferenciación
[dy_dw, dy_db] = tape.gradient(y, [w, b])
[dl_dw, dl_db] = tape.gradient(loss, [w, b])
```

```python
#El gradiente con respecto a una variable tiene el shape de la variable
print(w.shape)
print(dy_dw.shape)
print(dl_dw.shape)

print(b.shape)
print(dy_db.shape)
print(dl_db.shape)
```

```python
print("dy/dw = "+str(dy_dw.numpy()))
print("dy/db= "+str(dy_db.numpy()))
```

```python
print("dl/dw = "+str(dl_dw.numpy()))
print("dl/db= "+str(dl_db.numpy()))
```

```python
#Para comprobar, siempre podemos checar las variables que se están observando en tape para diferenciar
[var.name for var in tape.watched_variables()]
```

#### Otro ejemplo de diferenciación

```python
x = tf.linspace(-10.0, 10.0, 200+1) #Lista igual a la de numpy. En el contexto de tf, esto hace que x sea una constante (no diferenciable)

with tf.GradientTape() as tape:
  tape.watch(x) #Esto le dice a tape que ahora sí tome a x como una variable diferenciable
  y = tf.nn.sigmoid(x)

dy_dx = tape.gradient(y, x)
```

```python
plt.plot(x, y, label='y')
plt.plot(x, dy_dx, label='dy/dx')
plt.legend()
_ = plt.xlabel('x')
```

## Introduction to modules, layers, and models

```python
from datetime import datetime

%load_ext tensorboard
```

```python
#Modules y layers: Programación orientada a objetos en el contexto de deep learning

#Ejemplo de módulo escalar que opera en x, devolviendo a*x+b, con a entrenable y b no entrenable (constante)
class SimpleModule(tf.Module): #Definimos el objeto
  def __init__(self, name=None):
    super().__init__(name=name)
    self.a_variable = tf.Variable(5.0, name="train_me") #Las variables tienen (como siempre) valores predefinidos
    self.non_trainable_variable = tf.Variable(5.0, trainable=False, name="do_not_train_me")
  def __call__(self, x):
    return self.a_variable * x + self.non_trainable_variable #Definimos la operación entre las variables

#Creamos un módulo con la estructura del objeto
simple_module = SimpleModule(name="simple")

#Evaluamos al módulo en x=5
simple_module(tf.constant(5.0)) # 5*5+5 = 30
```

```python
# All trainable variables
print("trainable variables:", simple_module.trainable_variables)
# Every variable
print("all variables:", simple_module.variables)
```

#### Ejemplo de modelo lineal de dos capas

```python
#Primero una capa lineal densa (es decir, multilineal)
class Dense(tf.Module):
  def __init__(self, in_features, out_features, name=None): #features es el número de razgos de entrada/salida
    super().__init__(name=name)
    self.w = tf.Variable(
      tf.random.normal([in_features, out_features]), name='w') #Matriz de tamaño (in_features,out_features) con valores iniciales random
    self.b = tf.Variable(tf.zeros([out_features]), name='b') #Vector de tamaño (out_features) con valores iniciales cero
  def __call__(self, x):
    y = tf.matmul(x, self.w) + self.b #Operación multilineal
    return tf.nn.relu(y) #"Rectified Lineal Unit", función de activación que resulta en 0 para x<0 y x para x>0
```

```python
#El modelo completo, que crea dos instancias de capas y las aplica
class SequentialModule(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)
    
    #Creación de las dos capas
    self.dense_1 = Dense(in_features=3, out_features=3)
    self.dense_2 = Dense(in_features=3, out_features=2)

  def __call__(self, x): #Aplicación de las capas. En total entran 3 razgos y salen 2 razgos
    x = self.dense_1(x)
    return self.dense_2(x)

# You have made a model!
my_model = SequentialModule(name="the_model")

# Call it, with random results
print("Model results:", my_model(tf.constant([[2.0, 2.0, 2.0]])))
```

```python
#Todos los módulos se guardan de manera recursiva dentro del modelo (igual que en programación orientada a objetos)
print("Submodules:", my_model.submodules)
```

```python
#Igual para las variables
for var in my_model.variables:
  print(var, "\n")
```

```python
#Ejemplo de módulo al cual no se le tiene que especificar la dimensión de los razgos de entrada, 
#los deduce en la primera vez que lo llamas

class FlexibleDenseModule(tf.Module):
  # Note: No need for `in+features`
  def __init__(self, out_features, name=None):
    super().__init__(name=name)
    self.is_built = False
    self.out_features = out_features

  def __call__(self, x):
    # Create variables on first call.
    if not self.is_built:
      self.w = tf.Variable(
        tf.random.normal([x.shape[-1], self.out_features]), name='w')
      self.b = tf.Variable(tf.zeros([self.out_features]), name='b')
      self.is_built = True

    y = tf.matmul(x, self.w) + self.b
    return tf.nn.relu(y)
```

```python
# Used in a module. Efectivamente es el mismo modelo que el anterior, sólo que ahora los módulos son mas flexibles
class MySequentialModule(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)

    self.dense_1 = FlexibleDenseModule(out_features=3)
    self.dense_2 = FlexibleDenseModule(out_features=2)

  def __call__(self, x):
    x = self.dense_1(x)
    return self.dense_2(x)

my_model = MySequentialModule(name="the_model")
print("Model results:", my_model(tf.constant([[2.0, 2.0, 2.0]])))
```

#### Guardando los pesos

```python
#Podemos guardar los valores de los pesos
chkp_path = "my_checkpoint"
checkpoint = tf.train.Checkpoint(model=my_model)
checkpoint.write(chkp_path)

#Se crean dos archivos:
#Uno contiene los valores de los pesos como tales, otro los indíces de los pesos (metadata)
```

```python
#Podemos checar el checkpoint para ver que las variables sí se guardaron
tf.train.list_variables(chkp_path)
```

```python
#Podemos cargar el modelo a partir de ese archivo
new_model = MySequentialModule()
new_checkpoint = tf.train.Checkpoint(model=new_model)
new_checkpoint.restore("my_checkpoint")

# Should be the same result as above
new_model(tf.constant([[2.0, 2.0, 2.0]]))
```

## Basic Trainging Loops

```python
#Aquí se juntan todas las nociones de tensores, variables, "gradient tape" y módulos.
#Se mostrará primero un ajuste simple a un modelo lineal x*W+b

# The actual line
TRUE_W = 3.0
TRUE_B = 2.0

NUM_EXAMPLES = 1000

# A vector of random x values
x = tf.random.normal(shape=[NUM_EXAMPLES])

# Generate some noise
noise = tf.random.normal(shape=[NUM_EXAMPLES])

# Calculate y
y = x * TRUE_W + TRUE_B + noise
```

```python
# Plot all the data
#En la práctica se usan batches. Aquí los datos son tan pocos que se tomarán todos como un batch

plt.scatter(x, y, c="b")
plt.show()
```

```python
#Definimos el modelo lineal simple a ajustar
class MyModel(tf.Module):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    # Initialize the weights to `5.0` and the bias to `0.0`
    # In practice, these should be randomly initialized
    self.w = tf.Variable(5.0)
    self.b = tf.Variable(0.0)

  def __call__(self, x):
    return self.w * x + self.b

model = MyModel()

# List the variables tf.modules's built-in variable aggregation.
print("Variables:", model.variables)

# Verify the model works
model(3.0).numpy() == 15.0
```

```python
#Definimos la función de costo (mínimos cuadrados)
# This computes a single loss value for an entire batch
def loss(target_y, predicted_y):
  return tf.reduce_mean(tf.square(target_y - predicted_y)) #Promedio de los cuadrados
```

```python
#Predicciones iniciales (antes de entrenar)
plt.scatter(x, y, c="b")
plt.scatter(x, model(x), c="r")
plt.show()

print("Current loss: %1.6f" % loss(model(x), y).numpy())
```

#### Definición de un loop de entrenamiento
En general los loops consisten en

-Enviar un batch de inputs para generar outputs con el modelo.
-Calcular la función de costo de esos outputs.
-Usar "gradient tape" para encontrar los gradientes.
-Optimizar las variables con esos gradientes.

Aquí se utilizará el método de gradiente descendiente.

```python
# Given a callable model, inputs, outputs, and a learning rate...
def train(model, x, y, learning_rate):

  with tf.GradientTape() as t:
    # Trainable variables are automatically tracked by GradientTape
    current_loss = loss(y, model(x)) #En model(x) se sintetizan las relaciones entre los parámetros entrenables W y b, y la función de costo, a partir de los cuales se calcula su gradiente

  # Use GradientTape to calculate the gradients with respect to W and b
  dw, db = t.gradient(current_loss, [model.w, model.b])

  # Subtract the gradient scaled by the learning rate
  model.w.assign_sub(learning_rate * dw)
  model.b.assign_sub(learning_rate * db)
```

```python
#En el entrenamiento podemos ver como los parámetros evolucionan
model = MyModel()

# Collect the history of W-values and b-values to plot later
Ws, bs = [], []
epochs = range(10)

# Define a training loop
def training_loop(model, x, y):

  for epoch in epochs:
    # Update the model with the single giant batch
    train(model, x, y, learning_rate=0.1)

    # Track this before I update
    Ws.append(model.w.numpy())
    bs.append(model.b.numpy())
    current_loss = loss(y, model(x))

    print("Epoch %2d: W=%1.2f b=%1.2f, loss=%2.5f" %(epoch, Ws[-1], bs[-1], current_loss))
```

```python
print("Starting: W=%1.2f b=%1.2f, loss=%2.5f" %(model.w.numpy(), model.b.numpy(), loss(y, model(x))))

# Do the training
training_loop(model, x, y)

# Plot it
plt.plot(epochs, Ws, "r",
         epochs, bs, "b")

plt.plot([TRUE_W] * len(epochs), "r--",
         [TRUE_B] * len(epochs), "b--")

plt.legend(["W", "b", "True W", "True b"])
plt.show()
```

```python
# Visualize how the trained model performs
plt.scatter(x, y, c="b")
plt.scatter(x, model(x), c="r")
plt.show()

print("Current loss: %1.6f" % loss(model(x), y).numpy())
```
