```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

#### Definición de modelos y propagación de incertidumbres

```python
lambda_real = 532e-7 #cm
slambda_real = 0.5e-7 #cm

def y(x,M): #Modelo lineal general para ajustar
    return M*x

def sy(x,sx,M,sM):
    return np.sqrt((M*sx)**2+(x*sM)**2)

def L_model(D,d,lamb): #Modelo 
    n=1
    a=d/(n*lamb)
    M=1/np.sqrt(a**2-1)
    return M*D

def sL_model(D,sD,d,sd,lamb,slamb):
    n=1
    a=d/(n*lamb)
    M=1/np.sqrt(a**2-1)
    sa = a*np.sqrt((sd/d)**2+(slamb/lamb)**2)
    return M*np.sqrt(sD**2+M*(D*a*sa)**2)
```

```python
i=3
Datos = np.loadtxt("Datos"+str(i)+".txt",skiprows=1).T
Datos = Datos[ :, Datos[0].argsort()] #Ordena datos

num = len(Datos[0])
#num=6
ind = np.arange(0,num)

D = Datos[0][ind]
L1 = Datos[1][ind]
L2 = Datos[2][ind]
sD = Datos[3][ind]

#Modificaciones por grosor de CD o DVD
grosor = [0.15,0.15,0.9][i-1] #cm
sgrosor = [0.05,0.05,0.15][i-1] #cm

D = D-grosor
sD = np.sqrt(sD**2+sgrosor**2)

L = (L1+L2)/2 #Tomamos el promedio entre las dos distancias
sL = np.sqrt(0.5**2*(L1-L)**2+0.5**2*(L2-L)**2) #El error es la desviación estándar
```

### Ajuste lineal

```python
#Experimento 1 (fallido)
popt, pcov = curve_fit(y, xdata = D, ydata = L, sigma = sL)
M_fit = popt[0]
sM_fit = np.sqrt(pcov[0][0])

a_fit=np.sqrt(1/M_fit**2+1)
sa_fit=sM_fit/(a_fit*M_fit**3)

d_fit = lambda_real*a_fit*1e4 #cambiamos a mum
sd_fit = np.sqrt((lambda_real*sa_fit)**2+(a_fit*slambda_real)**2)*1e4 #cambiamos a mum

print("M_fit = "+str(M_fit)+" +- "+str(sM_fit))
print("a_fit = "+str(a_fit)+" +- "+str(sa_fit))
print("d_fit = "+str(d_fit)+" +- "+str(sd_fit)+" \mu m")
```

```python
#Experimento 2
popt, pcov = curve_fit(y, xdata = D, ydata = L, sigma = sL)
M_fit = popt[0]
sM_fit = np.sqrt(pcov[0][0])

a_fit=np.sqrt(1/M_fit**2+1)
sa_fit=sM_fit/(a_fit*M_fit**3)

d_real = 1.6e-4 #Para CD-R en cm
sd_real = 0.05e-4

#d_fit suponiendo lambda conocida
d_fit = lambda_real*a_fit*1e4 #cambiamos a mum
sd_fit = np.sqrt((lambda_real*sa_fit)**2+(a_fit*slambda_real)**2)*1e4 #cambiamos a mum
relerr_d = 100*np.abs(d_fit*1e-4-d_real)/d_real #d_fit cambia a cm para esta formula

#lambda_fit suponiendo d conocida
lambda_fit = d_real/a_fit*1e7 #Cambiamos a nm
slambda_fit = np.sqrt((sd_real/a_fit)**2+(d_real*sa_fit/a_fit**2)**2)*1e7 #cambiamos a nm
relerr_lambda = 100*np.abs(lambda_fit*1e-7-lambda_real)/lambda_real #lambda_fit cambia a cm para esta formula

print("M_fit = "+str(M_fit)+" +- "+str(sM_fit))
print("a_fit = "+str(a_fit)+" +- "+str(sa_fit))
print("d_fit = "+str(d_fit)+" +- "+str(sd_fit)+" \mu m, error relativo = " +str(relerr_d)+"%")
print("lambda_fit = "+str(lambda_fit)+" +- "+str(slambda_fit)+" nm, error relativo = " +str(relerr_lambda)+"%")
```

```python
#Experimento 3
popt, pcov = curve_fit(y, xdata = D, ydata = L, sigma = sL)
M_fit = popt[0]
sM_fit = np.sqrt(pcov[0][0])

a_fit=np.sqrt(1/M_fit**2+1)
sa_fit=sM_fit/(a_fit*M_fit**3)

d_real = 0.74e-4 #Para DVD en cm
sd_real = 0.01e-4

#d_fit suponiendo lambda conocida
d_fit = lambda_real*a_fit*1e4 #cambiamos a mum
sd_fit = np.sqrt((lambda_real*sa_fit)**2+(a_fit*slambda_real)**2)*1e4 #cambiamos a mum
relerr_d = 100*np.abs(d_fit*1e-4-d_real)/d_real #d_fit cambia a cm para esta formula

#lambda_fit suponiendo d conocida
lambda_fit = d_real/a_fit*1e7 #Cambiamos a nm
slambda_fit = np.sqrt((sd_real/a_fit)**2+(d_real*sa_fit/a_fit**2)**2)*1e7 #cambiamos a nm
relerr_lambda = 100*np.abs(lambda_fit*1e-7-lambda_real)/lambda_real #lambda_fit cambia a cm para esta formula

print("M_fit = "+str(M_fit)+" +- "+str(sM_fit))
print("a_fit = "+str(a_fit)+" +- "+str(sa_fit))
print("d_fit = "+str(d_fit)+" +- "+str(sd_fit)+" \mu m, error relativo = " +str(relerr_d)+"%")
print("lambda_fit = "+str(lambda_fit)+" +- "+str(slambda_fit)+" nm, error relativo = " +str(relerr_lambda)+"%")
```

### Gráfica

```python
D_cont = np.linspace(0,D.max()+3,100)

plt.figure(figsize=(8,6))
plt.errorbar(D,L,sL,sD,"r8",label="Promedios")
plt.errorbar(D,L1,0.05,sD,"bx",alpha=0.8,label="Derechos")
plt.errorbar(D,L2,0.05,sD,"mx",alpha=0.8,label="Izquierdos")
y_fit=y(D_cont,M_fit)
plt.plot(D_cont,y_fit,"g--",label="Ajuste")
sy_fit = sy(D_cont,0,M_fit,sM_fit)
plt.fill_between(D_cont,y_fit+sy_fit,y_fit-sy_fit,color=(0.7,0.9,0.7))
Titulo = [r"Relación $L_1 - D$ y su ajuste usando CD-R.",r"Relación $L_1 - D$ y su ajuste usando CD-R.",r"Relación $L_1 - D$ y su ajuste usando DVD."]
plt.title(Titulo[i-1])
plt.xlabel(r"$D[cm]$")
plt.ylabel(r"$L_1[cm]$")
plt.xlim(0,D.max()+3)
POSX = [0.1,20,20] #Datos2, Datos3
POSY = [0.1,1,3] #Datos2, Datos3
posx , posy = POSX[i-1], POSY[i-1]
props=dict(boxstyle='square', facecolor='white', alpha=1)
plt.text(posx, posy, r"Ajuste con $M_1$ = "+"{:.3f}".format(M_fit)+"$\pm$"+"{:.3f}".format(sM_fit),
         fontsize=14, bbox=props)
plt.legend()
plt.grid()
plt.show()
```
