```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
#Constantes en MKS
hbar = 6.6E-34
m = 9.1E-31 #Masa del electrón
VF = 10**6 #Velocidad de Fermi
```

```python
#Grafeno
def DG(E):
    return E/(np.pi*(hbar*VF)**2)

#Dispersión cuadrática en:
#3D
def D3(E):
    return m**(3/2)/(4*np.pi**2*hbar**3)*np.sqrt(E)
#2D
def D2(E):
    return m/(2*np.pi*hbar**2)*E/E
#1D
def D1(E):
    return m**(1/2)/(np.pi*hbar*np.sqrt(E))
```

```python
E = np.linspace(0.001,50,1000)

fig = plt.figure(figsize=(6,6))

#Factores de normalización para que tengan las mismas proporciones
plt.grid()
plt.plot(E,DG(E)/10**5,label='DG(E)')
plt.legend()

plt.plot(E,D3(E)/10**3,label='D3(E)')
plt.legend()

plt.plot(E,D2(E)*10**15,label='D2(E)')
plt.legend()

plt.plot(E,D1(E)*10**32,label='D1(E)')
plt.legend()

plt.show()
```
