### Resolviendo un cubo Rubik con una mano robot

Los autores del artículo [**Solving Rubik’s Cube
with a Robot Hand**](https://openai.com/blog/solving-rubiks-cube/) entrenaron algunas redes neuronales para resolver un cubo Rubik con una mano robot. Las redes fueron entrenadas solamente con simulaciones usando el mismo código de aprendizaje reforzado de **OpenAI Five** junto con una nueva técnica llamada *Automatic Domain Randomization* (ADR). El sistema puede manejar situaciones desconocidas durante el entrenamiento, provando que el aprendizaje reforzado no solo es una herramienta para tareass virtuales, sino que también puede resolver problemas del mundo real que requieren otras destrezas.

Me pareció un trabajo importante, ya que, como mencionan los autores, lograr que una mano robótica realice tareas complicadas como resolver el cubo es la base para la creación de robots para propósitos más generales.

Otra de las cosas interesantes que mencionan es su método *ADR* que automáticamente genera ambientes más parecidos a los del mundo real conforme la red aprende, incrementando la dificultad del ambiente.
