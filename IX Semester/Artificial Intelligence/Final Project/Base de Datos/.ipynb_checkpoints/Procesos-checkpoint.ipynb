```python
import numpy as np
from numpy.random import default_rng

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
rng = default_rng()
```

```python
n = 100
l_th = 32
l_s = 32
E_max = 14e12
```

```python
th0 = 2*np.pi*rng.random(l_th) - np.pi #genera angulos uniformes de -pi a pi
th0 = np.sort(th0)
th0
```

```python
x = np.arange(0,l_th)
plt.plot(x,th0, '.');
```

```python
#s0 = rng.exponential(E_max, l_s)
s0 = rng.normal(loc=rng.exponential(4*E_max**2,1), scale=E_max**2/10, size=l_s) #Distribución normal alrededor de un centro aleatorio

s0 = np.sort(s0)
s0
```

```python
x = np.arange(0,l_s)
plt.plot(x, s0, '.');
#plt.yscale('log')
```

```python
th, s = np.meshgrid(th0, s0)
th, s
```

```python
alpha = 1/137.035999084 #Constante de estructura fina
m_e = 0.511e6 #Masa de electrón
m_mu = 105.6e6 #Masa de muón
m_higgs = 125.10e6 #Masa del Higgs

def test_f(th, s):
    return s*np.sin(th)**2

#Electrón con muón (pair production)
def e_mu(th, s):
    E = np.sqrt(s/4) #s = 4E^2
    A = (m_mu/E)**2
    B = (E**2-m_mu**2)/(E**2-m_e**2)
    C = (m_e**2+m_mu**2)/E**2
    #Falta un factor de una raiz que es delicado porque es imaginario para energías prohibidas
    return (alpha/(4*E)**2)*np.sqrt(B)*(1+C+(1-A)*np.cos(th)**2)

#https://www.imperial.ac.uk/media/imperial-college/research-centres-and-groups/theoretical-physics/msc/current/qft/handouts/qftcrossection.pdf
def Yukawa(th, s):
    E = np.sqrt(s/4) #s = 4E^2
    m = m_higgs #Ponemos masa del higgs como la del bosón de interacción
    M = m_e #Masa del electrón como la del par de partículas que interactúan
    g = 64*np.pi**2*alpha #Que constante de acoplamiento ponemos???
    p = np.sqrt(E**2-M**2)
    A = 2*p**2*(1+np.cos(th))+m**2
    B = 2*p**2*(1-np.cos(th))+m**2
    return g**4/(128*np.pi**2*E)*(1/A+1/B)

#https://edu.itp.phys.ethz.ch/hs12/qft1/Chapter09.pdf (para los dos siguientes)
#Electrón con electrón (proceso de segundo orden)
def e_e(th, s):
    E = np.sqrt(s/4) #s = 4E^2
    m = m_e
    p = np.sqrt(E**2-m**2)
    A = (4*p**2+2*m**2)**2/(p**4*np.sin(th)**4)
    B = (8*p**4+12*m**2*p**2+3*m**4)/(p**4*np.sin(th)**4)
    return (alpha**2/(4*E**2))*(A-B+1)

#Electrón con positrón (Rutherford)
def e_p(th, s):
    #Este directo con las variables de Mandelstam en el CM
    m = m_e
    E = np.sqrt(s/4) #s = 4E^2
    p = np.sqrt(E**2-m**2)
    t = -2*p**2*(1-np.cos(th))
    u = -2*p**2*(1+np.cos(th))
    A = ((s-u)/t+(t-u)/s)**2
    B = 16*m**2*(5*m**2-2*u)/(s*t)
    return (alpha**2/(4*E)**2)*(A+B)
```

```python
#vro = test_f(th, s)
vro = e_mu(th, s)
#vro = Yukawa(th,s)
#vro = e_e(th,s)
#vro = e_p(th,s)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.

# Plot the surface.
surf = ax.plot_surface(th, s, vro, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(vro.min(), vro.max())
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
Ordmin = np.floor(np.log10(np.abs(np.min(vro)))) #Orden de magnitud mínimo
Ordmax = np.ceil(np.log10(np.abs(np.max(vro)))) #Orden de magnitud máximo
#RECORDAR QUE LOS VALORES EN (-linthresh,linthresh) SON EN ESCALA LINEAL
norm = cm.colors.SymLogNorm(linthresh=1E-30, linscale=0.3,vmin=1*10**Ordmin, vmax=1*10**Ordmax, base=10) #colores en escala logarítmica

fig = plt.figure()
ax = fig.add_subplot(111)
cset = ax.pcolormesh(th, s, vro, norm=norm, cmap='bone', shading='auto')
fig.colorbar(cset,extend='both')

plt.yscale('log')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$s[eV]$')
plt.grid()
plt.show()
```
