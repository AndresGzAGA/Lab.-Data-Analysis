# Elementos básicos de python y álgebra lineal

Un notebook de jupyter es una interfaz gráfica e interactiva a python. En python existen una gran variedad de librerías que nos pueden ayudar a llevar a cabo un sinnúmero de aplicaciones. Una de las más comunes es numpy, que incluye elementos básicos de álgebra lineal.

En la siguiente celda, se importa la librería numpy y se le asigna un alias "np" para llamarla. 

```python
import numpy as np
```

El elemento básico de numpy es el arreglo ("array"), que pude ser un vector visto como un vector o una matriz. [Aquí](http://webs.ucm.es/info/aocg/python/modulos_cientificos/numpy/index.html) puedes encontrar más información sobre numpy.

### Vectores
Construir un vector es muy intuitivo, pero si tienes dudas, puedes consultar la documentación de forma interactiva de la siguiente manera:

```python
np.array?
```

Ahora estás listx para crear tu primer vector (típicamente en física, usarás un vector tridimensional):

```python
a = np.array([1,1,1])
```

Como ya se mencionó, existen elementos básicos de algebra lineal incluidos en la librería numpy. Abajo, algunos de los ejemplos más sencillos:

```python
b = np.array([3,2,1])
```

```python
c = np.cross(a,b)
```

```python
print (c)
```

### Visualización

La librería para construir gráficas se llama pyplot y se encuentra dentro del paquete de visulización matplotlib. Lo podemos importar de forma análoga a como lo hicimos con numpy:

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

#### 2D
Empecemos con el caso 2D, y definamos dos vectores y visualicémoslos

```python
a1 = np.array([0.5, 1])
a2 = np.array([1, 0.5])
```

```python
plt.arrow?
```

```python
plt.arrow(0,0, a1[0],a1[1])
plt.arrow(0,0, a2[0],a2[1])
plt.show()
```

Esta función ya hace lo que queremos pero podría quedar mejor. Aquí unos tips para que la visualización quede mejor:

```python
plt.figure(figsize=(10,10))  # Especifica el tamaño de la imagen
plt.arrow(0,0, a1[0],a1[1], fc="b", ec="b", head_width=0.025, head_length=0.05 )
plt.arrow(0,0, a2[0],a2[1], fc="b", ec="b", head_width=0.025, head_length=0.05 )
plt.axis('equal') # Asegura que el tamaño en los dos ejes sea el mismo
plt.xlim((-0.25, 1.25)) # rango a graficar
plt.ylim((-0.25, 1.25)) # dominio a grafica
plt.axis('off')  # quita los ejes
plt.show()
```

Hay muchísimas posibilidades para hacer cosas con matplotlib. [Aquí](https://matplotlib.org/gallery/index.html) encontrarás una galería que muestra sus capacidades. 

#### 3D
De forma análoga podemos representar vectores en el espacio 3D usando la paquetería correspondiente:

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
a1 = np.array([0.5, 1,0.5])
a2 = np.array([1, 0.5,0.5])
a3 = np.array([0.5,0.5,1])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
Axes3D.plot3D(ax,[0,a1[0]],[0,a1[1]],[0,a1[2]],'k')
Axes3D.plot3D(ax,[0,a2[0]],[0,a2[1]],[0,a2[2]],'k')
Axes3D.plot3D(ax,[0,a3[0]],[0,a3[1]],[0,a3[2]],'k')
Axes3D.grid(ax,'off')
#Axes3D.axis(ax,'off')
plt.show()
```

### Matrices
Una matriz se puede expresar como un arreglo de dimensión más alta

```python
M = np.array([[1, 0.5, 0.5],
              [0.5, 1, 0.5],
              [0.5, 0.5, 1]])
```

En numpy, hay varias matrices típicas que se pueden construir en un solo comando:

```python
unos =np.ones((3,3))
print (unos)
```

```python
I = np.eye(3)
print (I)
```

```python
ceros = np.zeros((3,3))
print (ceros)
```

También existen las funciones elementales de álgebra lineal:

```python
np.matmul(M,I)
```

```python
np.linalg.inv(M)
```
