```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#DATOS 1
#Ignorados, la pila era muy pequeña
```

```python
#DATOS 2
#V0 = 1.443 #V
V0 = 1538 #mV
R = 9.85E3 #Ohm, feedback resistance
dR = 50
P1 = 983E3 #Ohm, Potenciometro, nominal
P2 = 8.3 #Ohm, Potenciometro, central
e = 1.602176634E-19 #Coulombs
kB_real = 1.380649E-23 #J/K
ek_real = e/kB_real
T = 273.15+17 #K
dT = 2
```

```python
#DATOS 3
V0 = 1538 #mV
R = 97.5 #Ohm, feedback resistance
dR = 1
#P1 = 983E3 #Ohm, Potenciometro, nominal
P2 = 10.11 #Ohm, Potenciometro, central
e = 1.602176634E-19 #Coulombs
kB_real = 1.380649E-23 #J/K
ek_real = e/kB_real
T = 273.15+20 #K
dT = 2
```

```python
def model_Ic(V1,I0,ek):
    return I0*np.exp(ek*V1/T) #I0*np.exp(e*V1/(kB*T))

def x(V1,T):
    return V1/T

def dx(V1,dV1,T,dT): #Incertidumbre de medición
    return 1/T*np.sqrt(dV1**2+V1**2*dT**2/(T**2))

def y(x,m,b): #Modelo lineal
    return m*x+b

def dy_fit(x,m,dm,b,db): #Incertidumbre de ajuste
    return np.sqrt(x**2*dm**2+db**2)

def dy_med(Ic,dIc): #Incertidumbre de medición
    return np.abs(dIc/Ic)
```

```python
i = 3 #i-ésimos Datos
R = [9.85E3,97.5][i-2]
```

```python
Datos = np.loadtxt("Datos"+str(i)+".txt",skiprows=1).T
Datos = Datos[ :, Datos[0].argsort()]*1E-3 #Ordena datos y convierte de mV a V

numi = [11,9] #Cantidad de datos menores que vamos a ajustar
num = numi[i-2]
#num = len(Datos[0])
ind = np.arange(0,num) #1 es si se quiere eliminar el dato mas bajo
ind_extra = np.arange(num,len(Datos[0]))

#Datos que sí se van a ajustar
V1 = Datos[0][ind] #Voltaje base-emisor
V2 = Datos[1][ind]
dV1 = Datos[2][ind]
dV2 = Datos[3][ind]
Ic = V2/R
dIc = np.sqrt(V2**2*dR**2/(R**4)+(dV2/R)**2)

#Datos extra medidos, pero que están fuera de la región lineal
V1_extra = Datos[0][ind_extra] #Voltaje base-emisor
V2_extra = Datos[1][ind_extra]
dV1_extra = Datos[2][ind_extra]
dV2_extra = Datos[3][ind_extra]
Ic_extra = V2_extra/R
dIc_extra = np.sqrt(V2_extra**2*dR**2/(R**4)+(dV2_extra/R)**2)

#Cambiamos datos para una función lineal
x_data = x(V1,T)
dx_data = dx(V1,dV1,T,dT)
y_data = np.log(Ic)
dy_data = dy_med(Ic,dIc)
```

### Ajuste lineal

```python
popt, pcov = curve_fit(y, xdata = x_data, ydata = y_data, sigma = dy_data)
ek_fit, b_fit = popt
dek_fit, db_fit = np.sqrt(pcov[0][0]), np.sqrt(pcov[1][1])
I0_fit = np.exp(b_fit)
dI0_fit = I0_fit*db_fit
print("e/kB ajustada: "+str(ek_fit)+" +- "+str(dek_fit))
print("e/kB real: " +str(ek_real))
print("Error porcentual con la real: "+str(np.abs(ek_fit-ek_real)/ek_fit))
```

```python
kB_fit = e/ek_fit
dkB_fit = e*dek_fit/(ek_fit**2)
print("kB ajustada: "+str(kB_fit)+" +- "+str(dkB_fit))
print("kB real: "+str(kB_real))
print("Error porcentual con la real: "+str(np.abs(kB_fit-kB_real)/kB_fit))
```

```python
#Aún con el ajuste a la recta, lo más práctico es graficar V1, Ic en escala logarítmica (exponencial en log scale)
V_cont = np.linspace(V1.min(),V1_extra.max(),100)
x_cont = V_cont/T

plt.figure(figsize=(8,6))
plt.errorbar(V1,Ic,dIc,dV1,"r+",label="Datos")
plt.errorbar(V1_extra,Ic_extra,dIc_extra,dV1_extra,"b+",label="Datos extra") #Datos extra que fueron medidos, pero no ajustados
plt.plot(V_cont,model_Ic(V_cont,I0_fit,ek_fit),"g--",label="Ajuste")
sy = dy_fit(x_cont,ek_fit,dek_fit,b_fit,db_fit)
plt.fill_between(V_cont,model_Ic(V_cont,I0_fit,ek_fit)*(1+sy),model_Ic(V_cont,I0_fit,ek_fit)*(1-sy),color=(0.7,0.9,0.7))
plt.title(r"Relación $V_1-I_c$ y su ajuste a los primeros "+str(num)+" datos. (R"+str(i-1)+" = "+str(int(R))+"$\Omega$)")
plt.yscale("log")
plt.xlabel(r"$V_1[V]$")
plt.ylabel(r"$I_c[A]$")
POSX = [0.13,0.10] #Datos2, Datos3
POSY = [0.5E-7,5E-6] #Datos2, Datos3
posx , posy = POSX[i-2], POSY[i-2]
props=dict(boxstyle='square', facecolor='white', alpha=1)
plt.text(posx, posy, r"Ajuste con $k_B$ = "+"{:.3e}".format(kB_fit)+"$\pm$"+"{:.3e}".format(dkB_fit)+"J/K",
         fontsize=14, bbox=props)
plt.legend()
plt.grid()
plt.show()
```

### Ajuste a la exponencial (en deshuso)

```python
popt, pcov = curve_fit(model_Ic, xdata = V1, ydata = Ic, sigma = dy_data)
I0_fit, ek_fit = popt
dI0_fit, dek_fit = pcov[0][0], pcov[1][1]
print("I0 ajustada: "+str(I0_fit))
print("e/kB ajustada: "+str(ek_fit)+" +- "+str(dek_fit))
print("e/kB real: " +str(ek_real))
print("Error porcentual con la real: "+str(np.abs(ek_fit-ek_real)/ek_fit))
```

```python
kB_fit = e/ek_fit
print("kB ajustada: "+str(kB_fit))
print("kB real: "+str(kB_real))
print("Error porcentual con la real: "+str(np.abs(kB_fit-kB_real)/kB_fit))
```
