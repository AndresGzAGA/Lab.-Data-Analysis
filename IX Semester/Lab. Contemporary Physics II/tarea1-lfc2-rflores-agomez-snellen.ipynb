```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
```

```python
N = 100000
t = 1e-7
T = 0.5
R = 1 - T
#g = rand.random(N)
```

```python
gt = np.float32(rand.random(N) <= T)
#gt
```

```python
ts = np.empty((2,N))
ts[1] = gt
ts[0] = np.linspace(0, N*t, N)
```

```python
plt.figure(figsize=(20,10))
plt.plot(ts[0],ts[1], 'x')
plt.show()
```
