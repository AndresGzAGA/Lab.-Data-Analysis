```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
%matplotlib inline
```

```python
#Constantes relevantes (en SI)
hbar = 6.62607004e-34
epsilon0 = 8.8541878176E-12
c_light = 299792458

Lambda = 810e-9 #Longitud de onda de emisión
k = 2*np.pi/Lambda
omega = c_light*k

dx = 2.5e-4 #Distancia entre rendijas
Dz = 0.57 #Distancia de la rendija a la pantalla
s = 5e-6 #Lado de cada pixel
A = 1e-5 #Ancho gaussiano del láser

Np = 3000 #numero de pixeles en cada lado
N = 200 #Promedio numero de fotones detectados por medición
```

```python
#GRÁFICA DEL PERFIL DE LA DISTRIBUCIÓN
plt.style.use('default')
x0 = np.linspace(-Np*s/2,Np*s/2,Np+1)
y0 = 50*s

r = np.sqrt(np.power(x0,2)+np.power(y0,2)+np.power(Dz,2)) #Distancia de la mitad de las rendijas al punto de observación
s1 = np.sqrt(np.power(x0-dx/2,2)+np.power(y0,2)+np.power(Dz,2)) #Distancia de rendija 1 al punto de observación
s2 = np.sqrt(np.power(x0+dx/2,2)+np.power(y0,2)+np.power(Dz,2)) #Distancia de rendija 2 al punto de observación

Phi = k*(s1-s2)
R = 100*s #Radio del volúmen de normalización (?)
f2r = hbar*omega/(2*epsilon0*4*np.pi*R)*1/r**2 #Función f(r) al cuadrado
I = f2r*(1+np.cos(Phi))*np.exp(-(np.power(x0,2)+np.power(y0,2))/A)
I = N*I/np.sum(I) #Modulamos para que en promedio se detecten N fotones

plt.figure(figsize=(6,4))
plt.plot(1e3*x0,I) #Graficamos en milimetros
plt.title('Perfil de la distribución de intensidad I(x)')
plt.ylabel('I(x)')
plt.xlabel('x [mm]')
plt.xlim(-1e3*Np*s/2, 1e3*Np*s/2)
plt.grid()
plt.show()
```

```python
#Calculo de la distribución en toda la placa
x = np.linspace(-Np*s/2,Np*s/2,Np+1)
y = np.linspace(-Np*s/2,Np*s/2,Np+1)
X, Y = np.meshgrid(x,y)

r = np.sqrt(np.power(X,2)+np.power(Y,2)+np.power(Dz,2)) #Distancia de la mitad de las rendijas al punto de observación
s1 = np.sqrt(np.power(X-dx/2,2)+np.power(Y,2)+np.power(Dz,2)) #Distancia de rendija 1 al punto de observación
s2 = np.sqrt(np.power(X+dx/2,2)+np.power(Y,2)+np.power(Dz,2)) #Distancia de rendija 2 al punto de observación

Phi = k*(s1-s2)

R = 100*s #Radio del volúmen de normalización (medio arbitrario, porque de todos modos se modula al final)
f2r = hbar*omega/(2*epsilon0*4*np.pi*R)*1/r**2 #Función f(r) al cuadrado
I = f2r*(1+np.cos(Phi))*np.exp(-(np.power(X,2)+np.power(Y,2))/A) #Modulación gaussiana para localizar el patrón
I = N*I/np.sum(I) #Modulamos para que en promedio haya N fotones por medición
```

```python
#plt.style.use('dark_background')
F_x, F_y = np.array([]), np.array([])
M = 2 #Numero de mediciones (para que se vea bonito en matplotlib notebook, tiene que ser par)

for m in range(M):
    fotones_init = rand.random((Np+1,Np+1)) #grid de numeros aleatorios
    i_y, i_x = np.asarray(np.where(fotones_init < I)) #Indices de fotones detectados
    f_x, f_y = 1e3*x[i_x], 1e3*y[i_y] #Coordenadas de fotones detectados en mm
    
    #Adjuntamos las coordenadas de la nueva iteración a las totales
    F_x = np.append(F_x,np.copy(f_x))
    F_y = np.append(F_y,np.copy(f_y))
    xy = np.vstack([F_x,F_y])
    z = gaussian_kde(xy)(xy) #Ajuste de color para densidad de puntos
    
    #plt.figure(figsize=(12,18))
    plt.figure(figsize=(12,4))
    plt.subplot(M/2, 2, m+1)
    plt.scatter(F_x,F_y,c=z,s=1)
    plt.xlim(-1e3*Np*s/2, 1e3*Np*s/2)
    plt.ylim(-1e3*Np*s/2, 1e3*Np*s/2)
    plt.title('Fotones en pantalla, medición '+str(m+1)+'.')
    plt.xlabel('x [mm]')
    plt.ylabel('y [mm]')
    ax = plt.gca()
    ax.set_facecolor((0,0,0))
    plt.show()
```
