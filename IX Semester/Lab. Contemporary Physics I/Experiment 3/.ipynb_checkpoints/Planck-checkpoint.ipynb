```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
h_real = 6.62607015e-34 #J*s
q = 1.602176634e-19 #C
c = 299792458 #m/s
#R = 102.7 #Ohm En los primeros dos experimentos
R = 100.2 #Ohm
sR = 0.05

def y(x,m,x0): #Modelo lineal general para ajustar
    return m*(x-x0)

def sy(x,sx,m,sm,x0,sx0):
    return np.sqrt(m**2*(sx**2+sx0**2)+sm**2*(x-x0)**2)

def V(nu,h,V0): #Modelo lineal para ajustar h, q exacta
    return h*nu/q+V0

def sV(nu,snu,h,sh,V0,sV0):
    return np.sqrt((nu*sh/q)**2+(h*snu/q)**2+sV0**2)
```

```python
L = ["Ro1a","Ro1b","Ro2","Ve1","Ve2","Ve3","Am1","Azul1","IR"]
Nombre = ["rojo 1a", "rojo 1b","rojo 2", "verde 1", "verde 2","verde 3" ,"amarillo 1", "azul 1","IR"]
lamb_rojo, slamb_rojo = (625+700)/2, (700-625)/2
lamb_verde, slamb_verde = (500+565)/2, (565-500)/2
lamb_amarillo, slamb_amarillo = (565+590)/2, (590-565)/2
lamb_cyan, slamb_cyan = (485+500)/2, (500-485)/2
lamb_azul, slamb_azul = (450+485)/2, (485-450)/2
lamb_IR, slamb_IR = 940, 940-760
Lamb = np.array([lamb_rojo,lamb_rojo,lamb_rojo,lamb_verde,lamb_verde,lamb_verde,lamb_amarillo,lamb_azul,lamb_IR])*1e-9 #m
sLamb = np.array([slamb_rojo,slamb_rojo,slamb_rojo,slamb_verde,slamb_verde,slamb_verde,slamb_amarillo,slamb_azul,slamb_IR])*1e-9 #m
V1sat = [0,0,0,0,0,0,0,0,0]
sV1sat = [0,0,0,0,0,0,0,0,0]
```

```python
i=5
Datos = np.loadtxt("Datos"+L[i]+".txt",skiprows=1).T
Datos = Datos[:, Datos[0].argsort()] #Ordena datos

LOW = [24,25,1,28,21,10,20,23,21]#indice menor
low = LOW[i] #indice menor
ind = np.arange(low,len(Datos[0])) #Datos a ajustar desde low hasta el máximo
ind_extra = np.arange(0,low)

V1 = Datos[1][ind]
V2 = Datos[0][ind]
sV1 = Datos[3][ind]
sV2 = Datos[2][ind]
Ic = V2/R
sIc = np.sqrt(V2**2*sR**2/(R**4)+(sV2/R)**2)

V1_extra = Datos[1][ind_extra]
V2_extra = Datos[0][ind_extra]
sV1_extra = Datos[3][ind_extra]
sV2_extra = Datos[2][ind_extra]
Ic_extra = V2_extra/R
sIc_extra = np.sqrt(V2_extra**2*sR**2/(R**4)+(sV2_extra/R)**2)
```

```python
#Rojo1a
popt, pcov = curve_fit(y, xdata = V1, ydata = Ic, sigma = sIc)
m_fit, x0_fit = popt
sm_fit, sx0_fit = np.sqrt(pcov[0][0]), np.sqrt(pcov[1][1])
print("m_fit = "+str(m_fit)+" +- "+str(sm_fit))
print("x0_fit = "+str(x0_fit)+" +- "+str(sx0_fit))
```

```python
V_cont = np.linspace(x0_fit,V1.max()+30,100)

plt.figure(figsize=(8,6))
plt.errorbar(x0_fit,0,0,sx0_fit,"mo",label=r"$V_{1,sat}$")
plt.errorbar(V1,Ic,sIc,sV1,"rx",label="Datos ajustados")
plt.errorbar(V1_extra,Ic_extra,sIc_extra,sV1_extra,"b.",label="Datos extra") #Datos extra que fueron medidos, pero no ajustados
y_fit = y(V_cont,m_fit,x0_fit)
plt.plot(V_cont,y_fit,"g--",label="Ajuste")
sy_fit = sy(V_cont,0,m_fit,sm_fit,x0_fit,sx0_fit)
plt.fill_between(V_cont,y_fit+sy_fit,y_fit-sy_fit,color=(0.7,0.9,0.7))
plt.title(r"Relación $I_d - V_1$ y su ajuste a los últimos "+str(len(Datos[0])-low)+" datos para el diodo "+ r"$\bf{"+Nombre[i]+ "}$.")
plt.xlabel(r"$V_1[mV]$")
plt.ylabel(r"$I_d[mA]$")
POSX = [0,1400,0,0,1600,2500,1550,2200,850] #DatosR1a, DatosR1b ,DatosR2
POSY = [5,40,4,4,35,1,35,20,40] #DatosR1a, DatosR1b, DatosR2
posx , posy = POSX[i], POSY[i]
props=dict(boxstyle='square', facecolor='white', alpha=1)
plt.text(posx, posy, r"Ajuste con raíz $V_{1,sat}$ = "+"{:.3f}".format(x0_fit*1e-3)+"$\pm$"+"{:.3f}".format(sx0_fit*1e-3)+"V.",
         fontsize=14, bbox=props)
plt.legend()
plt.grid()
plt.show()
```

```python
V1sat = np.array([1.6902297612485145,1.7250022946593577 ,1.8637044419946915,1.9155416946299397,2.666230174816681,1.8299562393131844,2.857198513987162,1.1328137087298692]) #V
sV1sat = np.array([0.011404106200614567,0.002182463407442696 ,0.005673913064320243,0.004115911566275859,.004659784290695643,0.004662941491587595,.023232154280806913,.004432573118138958]) #V
Lamb = np.array([lamb_rojo,lamb_rojo,lamb_verde,lamb_verde,lamb_verde,lamb_amarillo,lamb_azul,lamb_IR])*1e-9 #m
sLamb = np.array([slamb_rojo,slamb_rojo,slamb_verde,slamb_verde,slamb_verde,slamb_amarillo,slamb_azul,slamb_IR])*1e-9 #m
nu = c/Lamb #Hz
snu = nu*sLamb/Lamb #Hz
```

```python
#Ajuste a h
popt, pcov = curve_fit(V, xdata = nu, ydata = V1sat, sigma = sV1sat)
h_fit, V0_fit = popt
sh_fit, sV0_fit = np.sqrt(pcov[0][0]), np.sqrt(pcov[1][1])
print("h_fit = "+str(h_fit)+" +- "+str(sh_fit))
print("V0_fit = "+str(V0_fit)+" +- "+str(sV0_fit))
print("h_real = "+str(h_real))
print("Diferencia porcentual a la real: "+"{:.3f}".format(100*np.abs(h_fit-h_real)/h_real)+"%")
```

```python
numin, numax = 2e14, nu.max()+0.5e14
nu_cont = np.linspace(numin,numax,100)

plt.figure(figsize=(8,6))
plt.errorbar(nu,V1sat,sV1sat,snu,"rx",label="Datos ajustados")
y_fit = V(nu_cont,h_fit,V0_fit)
plt.plot(nu_cont,y_fit,"g--",label="Ajuste")
sy_fit = sV(nu_cont,0,h_fit,sh_fit,V0_fit,sV0_fit)
plt.fill_between(V_cont,y_fit+sy_fit,y_fit-sy_fit,color=(0.7,0.9,0.7))
plt.plot(nu_cont,V(nu_cont,h_real,0),"b:",label="Modelo esperado")
plt.title(r"Relación $V_{1,sat} - \nu$ y su ajuste a $h$.")
plt.xlim(numin,numax)
plt.xlabel(r"$\nu[Hz]$")
plt.ylabel(r"$V_{1,sat}[V]$")
posx , posy = 3e14, 0.6
props=dict(boxstyle='square', facecolor='white', alpha=1)
plt.text(posx, posy, r"Ajuste con $h$ = "+"{:.3e}".format(h_fit)+"$\pm$"+"{:.3e}".format(sh_fit)+r"$J\cdot s$.",
         fontsize=14, bbox=props)
plt.legend()
plt.grid()
plt.show()
```
