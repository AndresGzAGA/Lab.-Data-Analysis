  **Ejercicio 1: Generación de muestras y división 90% - 10%**

<font size=4, color = "black">
1. Genera un conjunto de 5000 muestras $(x_i, y_i)$ (tiempo y distancia) usando una pendiente $\beta = 15$ y una desviación $desv = 5$. Grafica las muestras.
    
2. Grafica el histograma para X y Y.
    
3. Divide el conjunto de muestras en $80$%, $10$% y $10$% (train, validation y test respectivamente). Y grafica las muestras para cada conjunto.
 
Puedes apoyarte de la función "generador_datos_simple(beta, m, desviacion)" del notebook "clase-3-vf.ipynb".

```python
def Muestra(beta,N,desv):
    """
    Genera una muestra de N números (x,y) con la relación y = beta*x
    con una desviación aleatoria dada por desv
    """
    x = random.rand(N)*10 #Genera x entre 0 y 10
    y = x*beta+random.randn(N)*desv
    return [x,y]
```

```python
def Muestra_plot(x,y,Legend):
    left, width = 0.1, 0.65
    bottom, height = 0.1, 0.65
    spacing = 0.005
    
    rect_scatter = [left, bottom, width, height]
    rect_histx = [left, bottom + height + spacing, width, 0.2]
    rect_histy = [left + width + spacing, bottom, 0.2, height]
    
    fig = plt.figure(figsize=(8, 8))
    
    ax = fig.add_axes(rect_scatter)
    ax_histx = fig.add_axes(rect_histx, sharex=ax)
    ax_histy = fig.add_axes(rect_histy, sharey=ax)
    
    ax.scatter(x,y,label=Legend)
    
    ax_histx.hist(x, bins=20)
    ax_histy.hist(y, bins=20, orientation='horizontal')
    
    ax_histx.tick_params(axis="x", labelbottom=False) #Quitar los números en los histogramas
    ax_histy.tick_params(axis="y", labelleft=False)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid()
    ax.legend()
    
    plt.show()
```

#### 1. y 2. Generación de muestras y gráfica

```python
beta,N,desv = 15,5000,5
xm,ym = Muestra(beta,N,desv)

Muestra_plot(xm,ym,"Muestra")
```

#### 3. Divisón del conjunto de muestras

```python
#Para sacar indices aleatoriamente sin repeptición
rng = random.default_rng()
ind_20 = rng.permutation(len(xm))[:int(len(xm)*0.2)] #indices del 20% que se va a separar
ind_10 = rng.permutation(len(ind_20))[:int(len(ind_20)*0.5)] #de ese 20% se saca la mitad para el 10%

#Indices de las 3 muestras
ind_train = np.delete(np.arange(0,len(xm)),ind_20) #indices del 80% de la lista
ind_val = ind_10 #indices de un 10% complementario
ind_test = np.delete(ind_20,ind_10) #indices del otro 10%

#Las 3 muestras
x_train, y_train = xm[ind_train], ym[ind_train]
x_val, y_val = xm[ind_val], ym[ind_val]
x_test, y_test = xm[ind_test], ym[ind_test]
```

```python
Muestra_plot(x_train,y_train,"Train")
```

```python
Muestra_plot(x_val,y_val,"Validation")
```

```python
Muestra_plot(x_test,y_test,"Test")
```

  **Ejercicio 2: Ajuste lineal y determinación de peso (weight) y bias óptimos**

<font size=4, color = "black">
1. Normaliza con una desviación estandar el conjunto de entrenamiento.
    
2. Normaliza el conjunto de validación y prueba usando el promedio y la desviación estándar calculados en el conjunto de entrenamiento.
    
3. Utilizando el conjunto de entrenamiento, realiza un ajuste empleando regresión lineal y la métrica del error mínimo cuadrado (MSE) usando el método de optimización "Descenso de gradiente".
    Nota: Recuerda que el ajuste se hace usando el conjunto de entrenamiento. Para cada iteración (epoch) calcula el residuo en el conjunto de entrenamiento y validación.

4. Grafica el costo (MSE en este caso) como función de la época (epoch) para los conjuntos train y validation.
    
5. Determina el residuo usando los pesos (weight) y bias óptimos encontrados en el paso 3 para el conjunto de prueba (test). 
 
Puedes apoyarte de la función "update_parameters(x, y, weight, bias, alfa, iteraciones)" del notebook "clase-3-vf.ipynb"

#### 1. Normalización a una desviación estandar del conjunto de entrenamiento

```python
x_mean = np.mean(x_train)
x_std = np.std(x_train)

x_train_norm = (x_train-x_mean)/(1.0*x_std)

y_mean = np.mean(y_train)
y_std = np.std(y_train)

y_train_norm = (y_train-y_mean)/(1.0*y_std)

Muestra_plot(x_train_norm,y_train_norm,"Entrenamiento normalizado")
```

#### 2. Normalización del conjunto de prueba y validación con el promedio y std del de entrenamiento

```python
x_val_norm = (x_val-x_mean)/(1.0*x_std)
y_val_norm = (y_val-y_mean)/(1.0*y_std)

x_test_norm = (x_test-x_mean)/(1.0*x_std)
y_test_norm = (y_test-y_mean)/(1.0*y_std)
```

#### 3. Ajuste utilizando regresión lineal, cálculo del resiudo para el conjunto de entrenamiento y validación para cada época

```python
def MSE(W, B, x, y, Deriv = False):
    """
    Calcula la métrica del error mínimo cuadrado MSE para los arrays pesos W y biases B (de igual tamaño)
    utilizando la muestra x,y.
    Si Derivatives == True también calcula su derivada con respecto a w y a b
    """
    adentro = (np.outer(x,W)+B).T-y #Array del término de adentro, con adentro[i][j] , i = (0,len(W)), j = (0,len(x))
    M = np.sum(adentro**2,axis=1)/len(x) #MSE[i] para cada peso de W
    if Deriv == False:
        return M
    else:
        dMSE_w = 2*np.sum(adentro*x,axis=1)/len(x)
        dMSE_b = 2*np.sum(adentro,axis=1)/len(x)
        return [M,dMSE_w,dMSE_b]
```

```python
def linear_fit(x, y, weight0, bias0, alfa, iteraciones):
    weights = []
    biases = []
    weight,bias = weight0, bias0
    
    for i in range(iteraciones):
        M = MSE(weight, bias, x, y, Deriv = True)
        r, d_w, d_b = M[0], M[1], M[2]
        
        weights.append(weight)
        biases.append(bias)
        
        # Gradiente descendente
        weight = weight - alfa*d_w
        bias = bias - alfa*d_b

    return weights, biases
```

```python
alfa = 0.02
iteraciones = 100
weight0 = random.random()
bias0 = random.random()

weights_fit, biases_fit = linear_fit(x_train_norm, y_train_norm, weight0, bias0, alfa, iteraciones)
w_fit, b_fit = weights_fit[-1], biases_fit[-1]

r_train = MSE(weights_fit, biases_fit, x_train_norm, y_train_norm, Deriv = False) #Residuos de train
r_val = MSE(weights_fit, biases_fit, x_val_norm, y_val_norm, Deriv = False) #Residuos de val
```

#### 3. Gráfica de MSE para train y validación

```python
epoch = np.arange(0,len(r_train))
fig = plt.figure(figsize=(10, 8))

plt.plot(epoch,r_train,"r+-",label="res de entrenamiento")
plt.plot(epoch,r_val,"g+-",label="res de validación")
plt.title("Residuos MSE como función de la época")
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.grid()
plt.legend()
plt.show()
```

#### 4. Residuos para conjunto de prueba

```python
r_test = MSE(weights_fit, biases_fit, x_test_norm, y_test_norm, Deriv = False) #Residuos de test
```

```python
fig = plt.figure(figsize=(10, 8))

plt.plot(epoch,r_test,"r+-",label="res de prueba")
plt.title("Residuos MSE como función de la época")
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.grid()
plt.legend()
plt.show()
```

  **Ejercicio 3: Inferencia**

<font size=4, color = "black">
Realiza una inferencia de la posición que tendrá una particula para los tiempos = [9.65, 6.76, 5.4, 9.4].

```python
def Inferencia(x,w,b,x_mean, x_std, y_mean, y_std):
    x = (x-x_mean)/x_std 
    y = w*x+b
    y = y*y_std+y_mean
    
    return y
```

```python
w_fit, b_fit = weights_fit[-1], biases_fit[-1]
x = np.array([9.65,6.76,5.4,9.4])

y_inferidas = Inferencia(x,w_fit,b_fit,x_mean, x_std, y_mean, y_std)
y_inferidas
```

```python
plt.figure(figsize=(8,8))

plt.plot(xm,ym,"b+",label="Muestra")
x = np.arange(0,10,0.1)
plt.plot(x,Inferencia(x,w_fit,b_fit,x_mean, x_std, y_mean, y_std),"r--",label="Ajuste")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica del ajuste lineal")

plt.show()
```

Repite el procedimineto de los ejercicio 1, 2 y 3 considerando que la depencia de Y con X está dada por: $𝐹(𝑥,𝑤,𝑏)=1.7159∗𝑡𝑎𝑛ℎ(𝑤𝑒𝑖𝑔ℎ𝑡∗𝑥+𝑏𝑖𝑎𝑠)$

```python
a_real = 1.7159
def F(x,w,b):
    return a_real*np.tanh(x*w+b)
```

```python
def MuestraH(w,b,N,desv):
    """
    Genera una muestra de N números (x,y) con la relación y = beta*x
    con una desviación aleatoria dada por desv
    """
    x = (random.rand(N)-0.5)*4 #Genera x entre -2 y 2
    y = F(x,w,b)+random.randn(N)*desv
    return [x,y]
```

```python
w_real,b_real,N,desv = 5, 0, 5000, 0.5
xm,ym = MuestraH(w_real,b_real,N,desv)

Muestra_plot(xm,ym,"Muestra")
```

```python
#Para sacar indices aleatoriamente sin repeptición
rng = random.default_rng()
ind_20 = rng.permutation(len(xm))[:int(len(xm)*0.2)] #indices del 20% que se va a separar
ind_10 = rng.permutation(len(ind_20))[:int(len(ind_20)*0.5)] #de ese 20% se saca la mitad para el 10%

#Indices de las 3 muestras
ind_train = np.delete(np.arange(0,len(xm)),ind_20) #indices del 80% de la lista
ind_val = ind_10 #indices de un 10% complementario
ind_test = np.delete(ind_20,ind_10) #indices del otro 10%

#Las 3 muestras
x_train, y_train = xm[ind_train], ym[ind_train]
x_val, y_val = xm[ind_val], ym[ind_val]
x_test, y_test = xm[ind_test], ym[ind_test]
```

```python
x_mean = np.mean(x_train)
x_std = np.std(x_train)

x_train_norm = (x_train-x_mean)/(1.0*x_std)

y_mean = np.mean(y_train)
y_std = np.std(y_train)

y_train_norm = (y_train-y_mean)/(1.0*y_std)

Muestra_plot(x_train_norm,y_train_norm,"Entrenamiento normalizado")
```

```python
x_val_norm = (x_val-x_mean)/(1.0*x_std)
y_val_norm = (y_val-y_mean)/(1.0*y_std)

x_test_norm = (x_test-x_mean)/(1.0*x_std)
y_test_norm = (y_test-y_mean)/(1.0*y_std)
```

```python
def MSEH(W, B, A, x, y, Deriv = False):
    """
    Calcula la métrica del error mínimo cuadrado MSE para los arrays pesos W y biases B (de igual tamaño)
    utilizando la muestra x,y con una relación y = 1.7159*np.tanh(x*w+b)
    Si Derivatives == True también calcula su derivada con respecto a w y a b
    """
    argumento = (np.outer(x,W)+B).T #Array del argumento de tanh, con argumento[i][j] , i = (0,len(W)), j = (0,len(x))
    A = np.meshgrid(A,A)[0].T #Bidimensionalizamos a A para poder multiplicarlo por la matriz con producto interno
    adentro = np.dot(A,np.tanh(argumento))-y
    M = np.sum(adentro**2,axis=1)/len(x) #MSE[i] para cada peso de W
    if Deriv == False:
        return M
    else:
        df = np.dot(A,np.power(np.cosh(argumento),-2))
        dMSE_w = 2*np.sum(adentro*df*x,axis=1)/len(x)
        dMSE_b = 2*np.sum(adentro*df,axis=1)/len(x)
        dMSE_a = 2*np.sum(adentro*np.tanh(argumento),axis=1)/len(x)
        return [M,dMSE_w,dMSE_b,dMSE_a]
```

```python
def tanh_fit(x, y, weight0, bias0, a0, alfa, iteraciones):
    weights = []
    biases = []
    ases = []
    weight,bias, ai = weight0, bias0, a0
    
    for i in range(iteraciones):
        M = MSEH(weight, bias, ai, x, y, Deriv = True)
        r, d_w, d_b, d_a = M[0][0], M[1][0], M[2][0], M[3][0]
        
        weights.append(weight)
        biases.append(bias)
        ases.append(ai)
        
        # Gradiente descendente
        weight = weight - alfa*d_w
        bias = bias - alfa*d_b
        ai = ai-alfa*d_a

    return weights, biases, ases
```

```python
alfa = 0.5
iteraciones = 500
weight0 = random.random()
bias0 = random.random()
a0 = random.random()

weights_fit, biases_fit, ases_fit = tanh_fit(x_train_norm, y_train_norm, weight0, bias0, a0, alfa, iteraciones)
w_fit, b_fit, a_fit = weights_fit[-1], biases_fit[-1], ases_fit[-1]

r_train = MSEH(weights_fit, biases_fit, ases_fit, x_train_norm, y_train_norm, Deriv = False) #Residuos de train
r_val = MSEH(weights_fit, biases_fit, ases_fit, x_val_norm, y_val_norm, Deriv = False) #Residuos de val
```

```python
epoch = np.arange(0,len(r_train))
fig = plt.figure(figsize=(10, 8))

plt.plot(epoch,r_train,"r+-",label="res de entrenamiento")
plt.plot(epoch,r_val,"g+-",label="res de validación")
plt.title("Residuos MSE como función de la época")
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.grid()
plt.legend()
plt.show()
```

```python
r_test = MSEH(weights_fit, biases_fit, ases_fit, x_test_norm, y_test_norm, Deriv = False) #Residuos de test
```

```python
fig = plt.figure(figsize=(10, 8))

plt.plot(epoch,r_test,"r+-",label="res de prueba")
plt.title("Residuos MSE como función de la época")
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.grid()
plt.legend()
plt.show()
```
