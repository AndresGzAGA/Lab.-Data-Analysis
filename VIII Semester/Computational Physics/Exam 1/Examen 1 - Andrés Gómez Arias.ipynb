### 1. Errores entre números decimales y números binarios

#### (2) Función frac_bin

```python
def frac_bin(D, cifras):
    '''
    Convierte el número Decimal D fraccionario (D<1) a un número fraccional binario con el número
    de cifras ingresado por el entero cifras.
    '''
    if D >= 1:
        print('Debes ingresar un número fraccionario menor a 1')
        return np.nan
    B = []
    for i in range(1,cifras+1):
        if D - 1/2**i <= 0: #Si es igual tendremos puras listas de 0, llegamos al redondeo exacto
            B.append(0)
        else:
            B.append(1)
            D -= 1/2**i
    return B
```

```python
frac_bin(0.1, cifras = 4)
```

```python
frac_bin(0.1, cifras = 10)
```

#### (3) Función frac_dec

```python
def frac_dec(B):
    '''
    Convierte el número fraccional binario expresado como lista B a decimal fraccionario.
    '''
    D = 0
    for i in range(len(B)):
        D += B[i]/2**(i+1)
    return D
```

```python
frac_dec([0,0,0,1])
```

```python
frac_dec([0,0,0,1,1,0,0,1,1,0])
```

#### (4) Errores porcentuales:

```python
d = np.sqrt(2)-1-frac_dec([0,1,1,0,1,0,1])
print(abs(d)/np.sqrt(2))
```

```python
d = np.pi-1*2-1-frac_dec([0,0,1,0,0,1,0,0,0,1])
print(abs(d)/np.pi)
```

### 2. Desarrollo de Métodos numéricos

(1) Método de la secante:

Éste método consiste en tomarte dos puntos del dominio de la función en los que ésta cambie de signo, de modo que puedas calcular la recta secante que une a los dos puntos y obtener la intersección de esta recta con y = 0. Si la función también es cero en ese punto, ya acabaste. Si es menor a cero, te tomas como siguiente iteración ese punto y el otro punto con f<0 de tu intervalo anterior. Si es mayor a cero, te tomas ese punto y el otro con f > 0. La convergencia del algoritmo es semilineal, pues expandiendo a primer orden todos los términos de la fórmula iterativa nos queda una relación asintótica $x_{k+1} = C x_{k}x_{k-1}$
Como el orden de convergencia se define como el número p tal que cuando k tiende a infinito $x_k = C x^p_{k-1}$, podemos resolver la relación asintótica sustituyendo la definición, nos quedará una ecuación $p = 1+1/p$ con solución $p = \frac{1+\sqrt5}{2}$. Justo porque aproximamos la función a una recta, despreciando términos cuadráticos, el error es cuadrático.

Los pros es que es sencillo de programar, no necesitas la derivada y siempre que haya un intervalo con cambio de signo y continua, converge. Los contras es que necesitas especificar un intervalo con cambio de signo, y para eso tienes que conocer al menos cualitativamente la función; si la función no es continua, no converge.

(2) Método de Newton:

Éste método consiste en tomarse un punto de la función y expandirla en serie de Taylor a orden 1,
de manera que obtienes la recta tangente en ese punto. Encuentras el cruze de esa recta con y = 0, 
y ese será tu siguiente punto de iteración. El funcionamiento de éste algoritmo se basa en que justo para puntos cercanos a una raíz, los términos de orden cuadrático en la serie de Taylor serán muy pequeños y la recta tangente será buena aproximación a la función cerca de la raíz.
Para la convergencia basta con aplicar la fórmula iterativa dos veces y tomar el término de menor orden, lo cual da una relación $x_{k} = C x^2_{k-1}$ debido a que en cada iteración se toman términos lineales, iterar dos veces cancela éstos términos y deja a siguiente orden los cuadráticos. De manera que su convergencia es de orden $p=2$, y su error igual es cuadrático.

Los pros son que también es sencillo de derivar, sólo tienes que poner un punto, que probablemente converga. Si la función tiene un polo o su derivada es pequeña cerca de la raíz, tendrá problemas para encontrarla, además necesitas conocer su derivada.

### 3. Inversa de la Tangente hiperbólica

```python
def arctanh(u ,E):
    '''
    Calcula arctanh(u) con un error menor a E << 1
    regresa [x, np.arctan(u)] donde x es el valor obtenido por éste algoritmo, y np.arctan(u) obtenido
    por el de numpy
    '''
    if abs(u) >= 1:
        print('Fuera del dominio')
        return [np.nan,np.nan]
    x = 0
    while True:
        xp = x+(u-np.tanh(x))*np.cosh(x)**2
        if abs(xp-x) < E:
            break
        x =xp
    return [x, np.arctanh(u)]
```
